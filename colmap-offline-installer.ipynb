# Ceres solver

````python
def install_ceres_solver():
    %cd /kaggle/working
    !cp -r /kaggle/input/colmap-offline-installer/ceres-solver-2.2.0 /kaggle/working
    !mkdir /kaggle/working/ceres-solver-2.2.0/build
    %cd /kaggle/working/ceres-solver-2.2.0/build
    !cmake ..
    !make -j$(nproc)
    !make install
    
    %cd /kaggle/working
    !rm -r /kaggle/working/ceres-solver-2.2.0

    print("ceres-solver-2.2.0 installed successfully")
````
